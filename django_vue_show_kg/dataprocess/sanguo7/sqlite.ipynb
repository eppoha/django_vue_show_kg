```python
def read_files_in_folder(folder_path):
    file_list = []
 
    for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        if os.path.isfile(file_path):
            file_list.append(file_path)
 
    return file_list
```

```python
pathls = read_files_in_folder(r'c:\Users\Administrator\Desktop\git\django_vue_show_kg\django_vue_show_kg\dataprocess\peopleresource')
pathls
```

```python
people_info = {}
for i in range(len(pathls)):
    
# for i in range(len(pathls)):
# for path in pathls:
    path = pathls[i]
    file = open(path, 'r+', encoding='utf-8')
    text = file.read()
    file.close()
    desc, text_all = text.split('\n\n')
    name = path.replace("c:\\Users\\Administrator\\Desktop\\git\\django_vue_show_kg\\django_vue_show_kg\\dataprocess\\peopleresource\\", '').split('.')[0].replace('\\', '')
    # people_info['name'] = name
    people_info[name] = {}
    people_info[name]['desc'] = desc
    people_info[name]['text_all'] = text_all
```
