```python
book_list = ["射雕英雄传","神雕侠侣","倚天屠龙记","三国"]
multi_flag = """百度百科是一部内容开放、自由的网络百科全书，旨在创造一个涵盖所有领域知识，服务所有互联网用户的中文知识性百科全书。在这里你可以参与词条编辑，分享贡献你的知识。"""
content = ""
soup = spider1(content)
# 获取简介
def get_desc(soup):
    try:
        description = soup.find('meta', attrs={'name': 'description'})
        print("===========人物描述============")
        desc = str(description).replace("<meta content=\"", "").replace("""" name="description"/>""","")
        print(desc)
        return desc
    except:
        print("=======当前人物爬取失败========")
        return "当前人物爬取失败"
# 获取完整介绍
def get_text_all(soup):
    try:
        # 获取所有标签
        all_tags = soup.find_all(True)
        text_all = ""
        for tags in all_tags:
            # print(i)
            class1 = tags.get('class')
            # print(type(class1))
            if "text_" in str(class1):
                # print((class1, tags))
                
                # print(tags.get_text())
                text_all += tags.get_text()
        # # text_all
        # text_Ygy92 = soup.findAll('span', attrs={'class': 'text_Ygy92'})
        # if text_Ygy92 == "" or text_Ygy92 == []:
        #     text_Ygy92 = soup.findAll('span',attrs={'class':"text_PiiCe"})
        # text_all = ""
        # for link in text_Ygy92:
        #     # print(link.get_text())  # 输出每个<a>标签的href属性值
        #     text_all += link.get_text()
        sub_word = "本词条缺少概述图，补充相关内容使词条更完整，还能快速升级，赶紧来编辑吧！"
        text_all = text_all.replace(sub_word, '')
        print("==============完整百科内容============")
        print(text_all)
        return text_all
    except:
        print("=======当前人物爬取失败========")
        return "当前人物爬取失败"
# 多词条处理
def get_multi_tag(soup):
    try:
        tags = soup.find_all('script')
        for tag in tags:
            # print(type(tag))
            tag_text = str(tag)
            if "window.PAGE_DATA" in tag_text and "lemmaId" in tag_text:
                tag_text = tag_text.replace("<script>", "").replace("</script>", "").replace("window.PAGE_DATA= ", "")
                print("text:",tag_text)
                tag_text = tag_text.split("{")
                # print("text:",text)
                for sub_tg in tag_text:
                    
                    # 得是所需剧本中
                    for book in book_list:
                        if book in sub_tg:
                            print("当前标签：",sub_tg)
                            # 按照逗号分割成一个个的键值对
                            new_str = sub_tg.split(',')
                            # print("new_str:",new_str)
                            tmp_str = new_str[0].split(':')
                            id = tmp_str[1]
                            
                            print("词条id：", id)
                            soup = spider2(content, id)
                            desc = get_desc(soup)
                            print("多词条挑选后的desc:", desc)
                            text_all = get_text_all(soup)
                            print("多词条挑选后的text_all:", text_all)
                            return desc, text_all
    except:
        print("=======当前人物爬取失败========")
        return "当前人物爬取失败", "当前人物爬取失败"
```

```python
# 爬取数据的主函数
def spider(name_ls:list, begin_index:int, end_index:int):
    # 保存失败数据
    failed_ls = set()
    for i in range(begin_index, end_index):
    # for i in range(len(shediaols)):
        print("=========当前索引为：****" + str(i) + "****=========")
        people = name_ls[i]
        print("===========当前人物：" + people + "===========")
        # 根据人物描述，先去看能否直接爬取
        soup = spider1(people)
        pre_desc = soup.find('meta', attrs={'name': 'description'})
        print("===========人物描述============")
        desc = str(pre_desc).replace("<meta content=\"", "").replace("""" name="description"/>""","")
        print(desc)
        # 完整百科
        text_all = ""
        # 没有词条，则直接跳过。
        if desc == "None":
            print("========当前人物===" + people + "===没有词条！===========")
            desc = "没有词条"
            failed_ls.add((i, people,"没有词条"))
        # 多个词条，需要找到并识别相符合的词条，并再次跳转至该相符合的词条去爬取。
        elif desc == multi_flag:
            print("========当前人物===" + people + "===有多个词条！===========")
            try:
                desc,text_all = get_multi_tag(soup)
            except:
                print("=======当前人物爬取失败：", people)
                desc,text_all = "当前人物多个词条爬取失败", "当前人物多个词条爬取失败"
                failed_ls.add((i, people,"当前人物多个词条爬取失败"))
        # data["简要"] = desc
        # 获取完整百科
        else:
            print("========正在爬取当前人物===" + people + "===的词条！===========")
            
            try:
                text_all = get_text_all(soup)
            except:
                print("=======当前人物完整百科爬取失败：", people)
                text_all = "当前人物完整百科爬取失败"
                failed_ls.add((i, people,"当前人物完整百科爬取失败"))

        # 保存路径
        pre_path = r"C:\Users\Administrator\Desktop\git\shediao\dataprocess\peopleresource"
        path = pre_path + "\\" + people + ".txt"
        # file = open(path, 'w+', encoding='utf-8')
        # file.write(text_all)
        # file.close()
        print("=========正在保存"+ people + "的数据=========")
        # with open(path, 'w+',encoding='utf-8') as file:
        #     json.dump(json_data, file, ensure_ascii=False)
        file = open(path, 'w+', encoding='utf-8')
        file.write(desc)
        file.write("\n\n")
        file.write(text_all)
        file.close()
        # 保存失败的数据
        
        failed_data_path = pre_path + "\\" + "failed_data.txt"
        file = open(failed_data_path, 'w+', encoding='utf-8')
        if failed_ls:
            for fd in failed_ls:
                file.writelines(str(fd))
        else:
            file.writelines("没有失败数据")
        file.close()
        print("=========保存成功，停顿15s=========")
        time.sleep(15)
    print("##################已经全部爬取完毕##################")
```
